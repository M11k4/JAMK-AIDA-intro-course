1 Read the data from the csv file and create a DataFrame from it.

2 Remove the following columns from the DataFrame: *Magnitude Seismic Stations*, *Azimuthal Gap*, *Horizontal Distance* and *Horizontal Error*.

3 Filter data from the DataFrame so that only rows matching the following criteria will be shown:
    - Source must be US
    - Magnitude value must be over 7.8

4 Create a new column to the tenth column position called Grade which will have values from the following formula: ("Magnitude" * "Depth") / "Root Mean Square". Results in this column should be presented with three decimals.

```python
# 1) Read the csv file and create a DataFrame from it.

import pandas as pd

quakes = pd.read_csv("data_files/earthquakes.csv",delimiter=",")

quakes.tail()
```

```python
# 2) Remove columns: Magnitude Seismic Stations, Azimuthal Gap, Horizontal Distance and Horizontal Error.

quakes.drop(["Magnitude Seismic Stations", "Azimuthal Gap", "Horizontal Distance", "Horizontal Error"],inplace=True,axis=1)
quakes.tail()
```

```python
# 3) Filter data from the DataFrame so that only rows matching the following criteria will be shown:
# - Source must be US
# - Magnitude value must be over 7.8
#- ID value must start with the string "USP0007"

#quakes[ (quakes.Source == 'US') & (quakes.Magnitude > 7.8) & (quakes.ID[:6] == 'USP0007') ]

#blake = quakes[quakes.Source == 'US']
#blake = quakes[quakes.Magnitude > 7.8]
#blake = quakes[(quakes.Source == 'US') & (quakes.Magnitude > 7.8)]

#blake = quakes.Magnitude  #.loc( lambda x : x > 7.8 )

#blake = quakes[ (quakes["Magnitude"] > 7.8) & (quakes["ID"][:6] == "USP0007") ]

#blake = quakes[ (quakes["ID"].isin(["USP0007"])) ]

#blake = quakes[ quakes["ID"].str.startswith('USP0007') == True ] WORKS!

#blake

quakes = quakes[ (quakes.Source == 'US') & (quakes.Magnitude > 7.8) & (quakes["ID"].str.startswith('USP0007') == True) ]

quakes
```

```python
# 4) Create a new column to the tenth column position called Grade which will have values from the following formula:
# ("Magnitude" * "Depth") / "Root Mean Square".
# Results in this column should be presented with three decimals.

quakes.insert(10, "Grade", round(quakes["Magnitude"]*quakes["Depth"]/quakes["Root Mean Square"],3), True)

quakes

# I wasn't sure if this was supposed to be done on the whole original df, or the filtered version from the previous task.
```
